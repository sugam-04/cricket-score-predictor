```python
import numpy as np
import pandas as pd
import os                      #The OS module in Python provides functions for creating and removing a directory (folder), fetching its contents, changing and identifying the current directory, etc.
from yaml import safe_load     #Yeh alag tarike ka file hai unlike csv aur excel
from tqdm import tqdm          #yeh basically time batata hai koi process kitne der mai hua
```

```python
file_name = []                                        #import karaya reason alag alag match ka alag file built rakha hai,there is no single file
for file in os.listdir('data'):
    file_name.append(os.path.join('data',file))
```

```python
file_name[0:3]
```

```python
final_df = pd.DataFrame()
counter = 1
for file in tqdm(file_name):                                #abhi kya kiya kyunki dataset csv format mai nahi tha toh isko json files mai normalize 
    with open(file,'r') as f:                               # tqdm jo hai woh batayega timing 
        df=pd.json_normalize(safe_load(f))
        df['match_id']= counter
        final_df = final_df.append(df)
        counter+=1
final_df        
```

```python
count = 1
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    delivery_df = delivery_df.append(loop_df)
```

```python
delivery_df
```

```python
def bowl(row):                                     #Apne pass batting team hai , bowling team nikalne ke liye simple function likh diya
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis=1)
```
